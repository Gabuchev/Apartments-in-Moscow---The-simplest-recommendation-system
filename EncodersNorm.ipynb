```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd 
```

```python
df = pd.read_csv(r'Data1.csv')
```

```python
df.head(10)
```

```python
Rdf=df["rooms"]
Mdf=df["metro_name"]
```

```python
labelencoder = LabelEncoder()
data_R= labelencoder.fit_transform(Rdf.values)
data_M = labelencoder.fit_transform(Mdf.values)
```

```python
Rooms_encoders = pd.DataFrame(data = data_R)
Rooms_encoders.columns=['Rooms_encoders']
Metro_encoders = pd.DataFrame(data = data_M)
Metro_encoders.columns=['Metro_encoders']
```

```python
Rooms_encoders
```

```python
Metro_encoders.head (10)
```

```python
#конкатенация
Data = pd.concat([df, Rooms_encoders,Metro_encoders], axis=1)
Data.head(2)
```

```python
Data=Data.drop(['metro_name',"rooms","address","id"],axis=1)#Удаляю столбец с предобработкой текста
```

```python
Data.head(2)
```

```python
Data.to_csv(r'Data2.csv',index=False)
```

```python
#Нормировка:

# импортирую preprocessing библиотеки sklearn
from sklearn.preprocessing import StandardScaler
 
# создаю объект этого класса
scaler = StandardScaler()
 
# привожу данные к единому масштабу 
Data_scaled = scaler.fit_transform(Data)
 
# заново создыю датафрейм с нормализованными значениями
Data_df = pd.DataFrame(Data_scaled, columns = Data.columns.values)
 
# первые пять строк
Data_df.head(-5)
```

```python
Data_df.to_csv(r'Data.csv',index=False)
```
